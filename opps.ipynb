## Types Of Constructor :

### 1. Default Constructor in Python

* The Python constructor which does not accept any parameter other than self is called as default constructor.
####  Example
* Let us define the constructor in the Employee class to initialize name and age as instance variables. We can then access these attributes through its object.

```python
class Employee:
   def __init__(self):
      self.name = "Ram"
      self.age = 24

e1 = Employee()
print ("Name: ",e1.name)
print ("age: ",e1.age)
```

### 2. Parameterized Constructor

* If a constructor is defined with multiple parameters along with self is called as parameterized constructor.
#### Example
* In this example, the __init__() constructor has two formal arguments. We declare Employee objects with different values −

```python
class Employee:
   def __init__(self, name, age):
      self.name = name
      self.age = age

e1 = Employee("Ram", 24)
e2 = Employee("Bharat", 25)

print ("Name: ",e1.name)
print ("age: ",e1.age)
print ("Name: ",e2.name)
print ("age: ",e2.age)
```

### Rules:

* Python constructor uses the def keyword and is always named __init__() with double underscore
* The first parameter must be self which refer to the instance and allows to access its attribute and method
* constructor are always define inside the class and are called automatically when an object is created
* if both default and paramaterized constructor exist python uses the default when no argument are passed and the paramiterized one when argument are provided.

## 5. Encapsulation

* Encapsulation in Python is used to bind variables and methods together inside a class and to restrict direct access to data.
* This is done by defining variable as private and providing public getter and setter method to access and medify them.
* for example : if i have a BankAccount class with private __balance variable. i can use method deposit() and withdraw() to modify the balance safely.

#### How does encapsulation work in python

* we declare private variables using double underscore (__) to hide them
* getter and setter methods are used to access or modify data safely
* only class itself can directly handle its private variables

#### Why do we need 

* we need encapsulation in python to hide the internal details of an object and protect data from direct access.
* encapsulation in python keeps internal data safe from accidental changes
* It allows controlled access through getter and setter method


```python
class Student:
    def __init__(self):
        self.roll_no=123
        self.__marks = 90      # private variable

    def show_marks(self):
        print(self.__marks)

s = Student()
print(s.roll_no)
s.show_marks()    # allowed
# s.__marks ❌ not allowed
```

```python
class Student:
    def __init__(self):
        self.__marks = 0        # private variable

    def get_marks(self):        # getter
        return self.__marks

    def set_marks(self, marks): # setter
        self.__marks = marks

# Object
s = Student()
s.set_marks(85)
print(s.get_marks())
```

## 5. Inheritance

* Inheritance is used to inherit the properties and behaviour of one class to another class.
* The class that inherit another class is called child class and class that gets inherited is called a base class or parent class.
  

This code shows inheritance where the User class is the parent and contains common properties and behavior such as name and login. The Student class is the child that inherits from User, so it can use these features without rewriting them. The Student class also adds its own behavior, study. When a Student object is created, it can use both the parent method (login) and its own method (study), making the code simple and reusable.

```python
class User:
    def __init__(self, name):
        self.name = name

    def login(self):
        print(self.name,"logged in")


class Student(User):
    def study(self):
        print(self.name,"is studying")


# object creation
s = Student("Ram")

# using inherited and own methods
s.login()   # from User
s.study()   # from Student
```

1. If the child class does not have its own constructor, Python automatically executes the parent class constructor when an object of the child class is created.

```python
# constructor example

class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")

class SmartPhone(Phone):
    pass

s=SmartPhone(20000, "Apple", 13)
s.buy()
```

2. If the child class has its own constructor, then Python will execute the child’s constructor instead of the parent’s.

```python
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

class SmartPhone(Phone):
    def __init__(self, os, ram):
        self.os = os
        self.ram = ram
        print ("Inside SmartPhone constructor")

s=SmartPhone("Android", 2)
# s.brand                    The SmartPhone(child) class has its own constructor, so Python does not run the Phone(parent) constructor. 
                             # This means brand is never initialized, and using s.brand causes an error.

print(s.os)
print(s.ram)
```

## 6. Super Keyword :

* Super() Keyword is used to access methods and constructor of parent class from child class.

#### It used to :

* To call parent class constructor
* To call parent class method
* To support multiple inheritance

#### 1. Calling Parent Class Constructor

```python
class Parent:
    def __init__(self, name):
        self.name = name
        print("Parent constructor")

class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)   # call parent constructor
        self.age = age
        print("Child constructor")

c = Child("Ram", 20)
print(c.name)  # Ram (from parent)
print(c.age)   # 20 (from child)
```

#### 2. Calling Parent Class method

```python
class Parent:
    def greet(self):
        print("Hello from Parent")

class Child(Parent):
    def greet(self):
        super().greet()   # call parent method
        print("Hello from Child")

c = Child()
c.greet()
```

### using super outside the class

* super() is used only inside a child class to call parent methods or constructor; it cannot be used outside the class.

```python
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")

class SmartPhone(Phone):
    def buy(self):
        print ("Buying a smartphone")
        # syntax to call parent ka buy method
        print(super().brand)

s=SmartPhone(20000, "Apple", 13)

super().buy()                     
```

### Access parent class variable using super()

* super() can only be used to call parent methods or constructors; it cannot directly access parent variables.

```python
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")

class SmartPhone(Phone):
    def buy(self):
        print ("Buying a smartphone")
        print(super().brand)            # try to access the parent class variable

s=SmartPhone(20000, "Apple", 13)

s.buy()
```

### Types of Inheritance

* Single Inheritance
* Multilevel Inheritance
* Hierarchical Inheritance
* Multiple Inheritance(Diamond Problem)
* Hybrid Inheritance

### 1. Single Inheritance
* parent class 
*       |
*       V
* Child class

* In single Inheritance a child class inherits attributes and method from only one parent class

```python
# Parent class
class Parent:
    def greet(self):
        print("Hello from Parent")

# Child class
class Child(Parent):
    def greet_child(self):
        print("Hello from Child")

# Create object
c = Child()
c.greet()       # inherited from Parent
c.greet_child() # own method of Child
```

### 2. Multiple Inheritance

* parent1     parent2 
*     |          |
*     V          V
*      Child class

```python
class Camera:
    def click(self):
        print("Camera takes photo")

class MusicPlayer:
    def play(self):
        print("Music is playing")

class SmartPhone(Camera, MusicPlayer):
    pass

s = SmartPhone()
s.click()   # from Camera
s.play()    # from MusicPlayer
```

### 3. Multilevel Inheritance

* Prent class
*      |
*      V
* Child class 1
*      |
*      V
* Child class 2

* In Multilevel inheritance a class is derived from another derived class
* There exist multiple layers of iheritance we can imagine it as a grandparent - parent - child relationship

```python
class Animal:
    def eat(self):
        print("Animal eats")

class Mammal(Animal):
    def walk(self):
        print("Mammal walks")

class Dog(Mammal):
    def bark(self):
        print("Dog barks")

d = Dog()

d.eat()    # from Animal
d.walk()   # from Mammal
d.bark()   # from Dog
```

### 4. Hierarchical Inheritance

* parent class
*  |--------|
*  V-------V
*  child  child

* In Hierarchical inheritance multiple child classes inherits from the same parent class 


```python
class Animal:
    def eat(self):
        print("Animal eats")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

class Cat(Animal):
    def meow(self):
        print("Cat meows")

d = Dog()
c = Cat()

d.eat()    # from Animal
d.bark()   # Dog's own method

c.eat()    # from Animal
c.meow()   # Cat's own method
```

### 4. Hybrid Inheritance

* Combination of two or more types of inheritance is called as hybrid inheritance.

```python
class A:
    def show(self):
        print("Class A")

class B(A):
    def show_b(self):
        print("Class B")

class C(A):
    def show_c(self):
        print("Class C")

class D(B, C):
    def show_d(self):
        print("Class D")

d = D()
d.show()    # from A
d.show_b()  # from B
d.show_c()  # from C
d.show_d()  # from D
```

## 7. Polymorphism:

#### Polymorphism means taking many forms.

* Method Overriding
* Method Overloading
* Operator Overloading

### 1. Method Overriding

* Method Overriding refers to defining a method in subclass with the same as a method in its superclass.

```python
class Parent:
    def show(self):
        print("Parent method")

class Child(Parent):
    def show(self):
        print("Child method")

c = Child()
c.show()   # Child method
```

### 2. Method Overloading

* Method overloading where a class can have multiple methods with same name but different parameters. To overload method, we must change the number of parameters.

```python
class Shape:

  def area(self,a,b=0):
    if b == 0:
      return 3.14*a*a
    else:
      return a*b

s = Shape()

print(s.area(2))
print(s.area(3,4))
```

### 2. Operator Overloading

* Operator overloading in Python means the same operator can do different things for different data types.
* Python built-in data types allow + operator can add numbers, join strings or merge lists and * operator can be used to repeat instances of a string.

```python
# + operator
print(10 + 5)          # adds numbers → 15
print("Hi" + " All")   # joins strings → Hi All
print([1, 2] + [3, 4]) # merges lists → [1, 2, 3, 4]

# * operator
print("Hi " * 3)       # repeats string → Hi Hi Hi 
print([1] * 4)         # repeats list → [1, 1, 1, 1]
```

## 8. Abstraction:

* Abstraction means hiding unnecessary details and showing only the important features of an object

### Create an abstract class:

* To create an abstract class in python it must inherits the ABC (abstract base class) that is defined in the ABC module.
* The class must have at least one abstarct method
* An abstract method is one which cannot be called but can be overriden. You need to declare with @abstractmethod decorator

### Abstract Method:

* Abstract method are method declaration without a body defined inside an abstract classes.
* They act as placeholders that force subclasses to provide their own specific implementation.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        print("Car starts with a key")

class Bike(Vehicle):
    def start(self):
        print("Bike starts with a button")

c = Car()
b = Bike()

c.start()
b.start()
```
