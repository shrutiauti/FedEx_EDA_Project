# 5. SQL Constraints?

* SQL constraints are rules applied to columns or tables in a database to limit type of data that can be inserted, updated or deleted

## Types Of Constrinst 
* 1. NOT NULL
* 2. UNIQUE
* 3. PRIMARY KEY
* 4. FOREIGN KEY
* 5. CHECK
* 6. DEFAULT

### 1. NOT NULL
* This constraint ensures that a column cannot have a NULL (empty) value.
* It means every row must contain a value in this column.
  
### 2. UNIQUE
* This constraint ensures that all values in a column are different from each other.
* No duplicate values are allowed in that column.
  
### 3. PRIMARY KEY
* It is used to uniquely identify each row in a table.
* It is combination of NOT NULL and UNIQUE constraints, so duplicate or empty values are not allowed.
* A table can have only one primary key.
  
### 4. FOREIGN KEY
* It is used to create a relationship between two tables.
* It refers to the primary key of another table and ensures valid data connection between tables.
  
### 5. CHECK
* This constraint ensures that the values in a column satisfy a specific condition.
* For example, age must be greater than 18.
  
### 6. DEFAULT
* It assigns a default value to a column if no value is provided during insertion.
* For example, if no city is given, it can automatically set it to "Pune".

```python
%%sql

CREATE TABLE Departments (
    dept_id INT PRIMARY KEY,
    dept_name VARCHAR(50) NOT NULL
);

CREATE TABLE Students (
    id INT PRIMARY KEY,                          
    name VARCHAR(50) NOT NULL,                   
    email VARCHAR(100) UNIQUE,                   
    age INT CHECK (age >= 18),                   
    city VARCHAR(50) DEFAULT 'Pune',             
    dept_id INT,                                 
    FOREIGN KEY (dept_id) REFERENCES Departments(dept_id)
);
```

## 1. WHERE 
* WHERE clause filters records, based on one or more conditions, so your query returns the records that match.

* It is used with SELECT, UPDATE, and DELETE statements.

* It works with comparison and logical operators.

* where is always used before GROUP BY caluse.

* Example : If I want to increase the salary of employees who have more than 10 years of experience, I use the WHERE clause to filter those specific rows, and then apply the UPDATE operation only on them.

```python
%%sql

SELECT *
FROM Stud
WHERE age > 20;
```

## 2. GROUP BY

* The group by caluse is used to group rows that have same values in specified column.

* it is useful when we want to summarize and analyze large data to extract meaningful insights

* it is used with aggregation function like count(), min(), max() etc. to perform calculation on each group of data

* Example:
If a company wants to find the average salary in each department, we use GROUP BY department and apply AVG(salary) to calculate the average salary for each department.

```python
%%sql

CREATE TABLE Employees (
    id INT,
    name VARCHAR(50),
    department VARCHAR(50),
    salary INT
);
```

```python
%%sql

SELECT department, COUNT(*) AS total_employees
FROM Employees
GROUP BY department;
```

## 3. Having 

* The HAVING clause in SQL is used to filter grouped records based on a condition.
  
* HAVING is always used after GROUP BY caluse.
  
* Example : “If each department should have at least 15 employees, and I want to find departments with less than 15 employees so we can hire more people, I first count employees in each department and then use HAVING to filter the result.”

```python
%%sql

SELECT department, COUNT(*) AS total_employees
FROM Employees
GROUP BY department
HAVING COUNT(*) > 1;
```

## 2. ORDER BY 

* The ORDER BY  caluse is used to sort data in either in ascending or descending order based on one or two columns.
* ORDER BY CLAUSE is specified after WHERE, GROUP BY and HAVING clause.

## 9. Joins 

* The Join clause is used to combine the rows from two or more tables, based on a common column between them.
* Joins are used when we want to retrieve data that is distributed across multiple tables, and information we need is not available in single table

### SYNTAX

- SELECT column_names
- FROM table1
- JOIN table2
- ON table1.column = table2.column;

### Types Of Join

1. INNER JOIN
2. LEFT JOIN (LEFT OUTER JOIN)
3. RIGHT JOIN (RIGHT OUTER JOIN)
4. FULL JOIN (FULL OUTER JOIN)
5. CROSS JOIN
6. SELF JOIN

## 1. INNER JOIN

* INNER JOIN returns only the records that have matching values in both tables.

* Example : Suppose we have a Customers table and an Orders table. If we are asked to display only those customers who have placed orders, we use an INNER JOIN. It returns only the matching records from both tables.Customers without orders will not appear in the result.

```python
%%sql

SELECT e.id, e.name, e.department, e.salary, d.dept_id
FROM Employees e
INNER JOIN Departments d
ON e.department = d.dept_name;
```

## 2. LEFT JOIN

* LEFT JOIN returns all records from the left table and matching ones from the right table.

* Example : Suppose we want to show all customers and their orders. Even if a customer hasn’t placed any order, they will still appear, and their order details will be shown as NULL.

* Suppose we want to display all customers along with their order details, even if some customers have not placed any orders. In this case, we use a LEFT JOIN where the Customers table is the main table. It returns all customers and matching orders if available. If a customer has no order, the order-related columns will show NULL.

* another example : We use Students and Fees tables to find students who have not paid fees. We LEFT JOIN Students with Fees, and the rows where fee details are NULL represent students with pending fees.

* We use Candidates and Interviews tables to find candidates who have not been called for interview. We LEFT JOIN Candidates with Interviews, and where interview details are NULL, it means those candidates are still not scheduled for any interview.


```python
%%sql

SELECT e.id, e.name, e.department, e.salary, d.dept_id
FROM  Departments d
LEFT JOIN Employees e
ON e.department = d.dept_name;
```

## 3. RIGHT JOIN

* RIGHT JOIN returns all records from the right table and matching ones from the left table.

* Suppose we have Employees and Departments tables and we want to show all departments and their employees. Even if a department has no employees, it will still appear, and employee details will be shown as NULL.

* Suppose we have Employees and Departments tables and we want to list all departments along with employees working in them, including departments that have no employees. In this case, we use a RIGHT JOIN. It returns all rows from the Departments table and matching employees from the Employees table. If a department has no employees, employee fields will show NULL.

```python
%%sql

SELECT e.id, e.name, e.department, e.salary, d.dept_id
FROM Employees e
RIGHT JOIN Departments d
ON e.department = d.dept_name;
```

## 4. FULL JOIN 

* FULL JOIN returns all records when there is a match in either the left or the right table.


```python
%%sql

SELECT e.id, e.name, e.department, e.salary, d.dept_id
FROM Employees e
full JOIN Departments d
ON e.department = d.dept_name;
```

## 5. SELF JOIN

* SELF JOIN is a join where a table is joined with itself to compare rows within the same table.

* Suppose we want to show each employee along with their manager. We join the same Employees table with itself to match employees with their managers.

* Suppose we have an Employees table where each employee has a manager_id column referring to another employee in the same table. If we want to display each employee along with their manager’s name, we use a SELF JOIN. Here, the same table is joined with itself to match employees with their respective managers.
  


```python
%%sql

SELECT e.name AS Employee, m.name AS Manager
FROM Employees e
INNER JOIN Employees m
ON e.manager_id = m.id;
```

## 6. CROSS JOIN

* CROSS JOIN returns the Cartesian product of the two tables, pairing each row from one table with all rows from the other.

* Suppose a coffee shop has 2 coffee types and 3 cup sizes. A CROSS JOIN produces all combinations like Espresso–Small, Espresso–Medium, Latte–Small, Latte–Medium, etc.

```python
%%sql 

SELECT e.id, e.name, e.department, e.salary, d.dept_id
FROM Employees e
CROSS JOIN Departments d;
```

## 11. WINDOW FUNCTION

* SQL window functions allow performing calculations across a set of rows that are related to the current row, without collapsing the result into a single value.

* perform the calculation on subset of data without losing the level of details of rows
  
* They are commonly used for tasks like aggregates, rankings, and running totals.
  
* Instead of reducing rows like GROUP BY, they return results for each row (row-by-row calculation).

### SYNTAX

* SELECT column_name1, 
*        window_function(column_name2) 
*        OVER ([PARTITION BY column_name3] [ORDER BY column_name4]) AS new_column
* FROM table_name;


* window function : perform a calculation within a window
* The OVER clause defines the “window” of rows for the calculation. It can:

* PARTITION BY: It divides the data into groups using PARTITION BY.
* ORDER BY: It specifies the order of rows within each group using ORDER BY.
* With this, functions such as SUM(), AVG(), ROW_NUMBER(), RANK() and DENSE_RANK() can be applied in a controlled way.

## Types of window function

### 1. ROW_NUMBER

* Assigns a unique number to each row in the result set.
* dont handle the ties

#### use cases:
* 

```python
%%sql

SELECT name, dept, salary,
       ROW_NUMBER() OVER (PARTITION BY dept ORDER BY salary DESC) AS row_num
FROM Employees;
```

## 2. RANK() 

* Assigns ranks to rows, skipping ranks for duplicates.
* handle the ties
* it leaves the gap in ranking

### use cases :
* top N analysis : analysis the top performers to do targeted marketing
* bottom n analysis : help analysis the underperformance to manage risks and to do optimizations
* assign unique ids : help to assign identifier for each row to help pagination
* identify the duplicates : identify and remove duplicate to improve data quality

```python
%%sql

select name,dept,salary,
RANK() OVER(PARTITION BY dept ORDER BY salary DESC)
FROM employees
```

## 3. DENSE_RANK()

* Assigns ranks to rows without skipping rank numbers for duplicates.
* it handles the ties

#### 1.  Q: Suppose we create a view on a table. If we later add a new column to the base table, will it automatically be added to the view?
* No, adding a new column to the base table does not automatically get added to the view. A view only includes the columns that were defined in its original SELECT query.If we want the new column in the view, we must update it manually using CREATE OR REPLACE VIEW.

#### 2. If we update records in a view, will it reflect in the base table?

* A: It depends on the type of view. If the view is made from a single table, then any update in the view will also change the original table. But if the view is complex, like using joins, GROUP BY, or calculations, then update is not allowed or it will not reflect in the base table.

#### 3. If a view contains DISTINCT, GROUP BY,WINDOW FUNCTION,or WITH CALUSE can it be updated?

* A: If the query inside a view contains DISTINCT, GROUP BY, WINDOW FUNCTION,or WITH CALUSE then the view cannot be updated. This is because these operations change or group the data, so SQL cannot directly map changes back to the original table. Such views are read-only, and we cannot perform INSERT, UPDATE, or DELETE operations on them.

```python
%%sql

WITH cte_total_amt1 AS (
    SELECT user_id,
           SUM(amount) AS total_amount
    FROM orders
    GROUP BY user_id
),

cte_avg_delitime AS (
    SELECT user_id,
           MAX(delivery_time) AS avg_deltime
    FROM orders
    GROUP BY user_id
),

cte_avg_delirating as (
    select user_id,
        MAX(delivery_rating) as avg_delirating
    from orders
    group by user_id
),

cte_order_rank as (
      select user_id,total_amount,
      rank() over(order by total_amount desc) as rank_orders
      from cte_total_amt1
),

cte_segment_orders as (
     select user_id,rank_orders,
      case when rank_orders = 1 then 'very good'
           when rank_orders = 2 then 'good'
           when rank_orders = 3 then 'ok ok '
           else 'average'
           end as segment_ordrs
      from cte_order_rank
)

SELECT 
    o.order_id,
    o.delivery_time,
    o.delivery_rating,
    c.total_amount,
    l.avg_deltime,
    r.avg_delirating,
    rk.rank_orders,
    so.segment_ordrs
FROM orders o
LEFT JOIN cte_total_amt1 c
    ON o.user_id = c.user_id
LEFT JOIN cte_avg_delitime l
    ON o.user_id = l.user_id
LEFT JOIN cte_avg_delirating r
    ON o.user_id = r.user_id
LEFT JOIN cte_order_rank rk
    ON o.user_id = rk.user_id
LEFT JOIN cte_segment_orders so
    ON o.user_id = so.user_id;
```

# 17. NORMALIZATION

* Normalization is a technique used to remove and reuce redundancy(Redundancy means repeating the same data or information again and again unnecessarily.)

* It divides large unstructure table into smaller related table and connect them using keys

* The main goal is to make database more efficient consistent

* There are different normal forms like 1NF, 2NF, & 3NF each remove different type of redundancy and dependancy.

## 1NF :

* Each column should have only one value (no multiple values in a single cell).
  
* Example : If a table has a column phone Numbers that store nultiple phone Number in a single cell . It violets 1NF.

## 2NF :

* Must be in 1NF
  
* Remove partial dependency , means every non-prime attribute must depends on the entire key not just part of it

* Example : For composit key (StudentID, CourseID) if the 'StudentName' depends only on StudentID and not entire key . move StudentName into seperate table where it depends only on StudentID

## 3NF :

* Must be in 2NF

* Remove transitive dependency

* Example : Consider table with (StudentID, CorseID, InstructorID) . If InstructorID depends on CorseID & CorseID depends on StudentID then InstructorID directly depends on StudentID. It Violets 3NF.

## BCNF (Boyce-Codd Normal Form)

* Every determinant must be a candidate key

* If table has dependancy (StudentID, CorseID) -> Instructor but neither StudentID nor CourseID is candidate key . then its violets BCNF

* Decompose the table so that each determinant is candidate key.


## 4NF 

* Remove multi-valued dependency

* Example : Consider a table where (StudentID, language, Hobby) are attributes . If Student can multiple hobbies and language. multivalued dependancy exist.To resolve this split the table into seperate table for language and hobbies.


## 5NF 

* Remove join dependency

* It ensure that every table is fully decompose into smaller tables that are logically connected without loosing information.

* 5NF splits the table into smaller tables in such a way that:
- No redundancy remains
- Only valid combinations exist
- Joining them gives correct data only

## DENORMALIZATION :

* It the processes of combining normalized table to improve performance at cost of introducing redundancy.

# 18. ACID 

## 1. Atomicity :

* Atomicity means a transaction is all-or-nothing either all its operations succeed, or none are applied. If any part fails, the entire transaction is rolled back to keep the database consistent.

* Commit: If the transaction is successful, the changes are permanently applied.

* Abort/Rollback: If the transaction fails, any changes made during the transaction are discarded.

* Example: Consider the following transaction T consisting of T1 and T2 : Transfer of $100 from account X to account Y .
If the transaction fails after completion of T1 but before completion of T2, the database would be left in an inconsistent state. With Atomicity, if any part of the transaction fails, the entire process is rolled back to its original state, and no partial changes are made.


## 2. Consistency :

* Consistency ensure that before and after transaction database remain valid.

* Example : Before transaction satrt and after transaction completed sum of the money should be remain.

* before transaction the total balance 700
* before transaction the total balance 700

## 3. Isolation :

* Isolation ensures that transactions run independently without affecting each other.
  
* Changes made by one transaction are not visible to others until they are committed.


## 4. 4. Durability:

* Durability ensures that once a transaction is committed, its changes are permanently saved, even if the system fails.

* Example: After successfully transferring money from Account A to Account B, the changes are stored on disk. Even if there is a crash immediately after the commit, the transfer details will still be intact when the system recovers, ensuring durability.

# 19. KEYS

## 1. Primary Key

* A primary key is a unique identifier for a record in a table.
* It ensures that each row has a distinct value and prevents duplicate records.
* A table can have only one primary key, and its values cannot be NULL.

## 2. Foreign Key

* A foreign key is a column or a set of columns in one table that establishes a relationship with the primary key of another table.

## 3. Unique Key

* A unique key ensures that all values in the column are distinct but allows one NULL value.
* Unlike the primary key, a table can have multiple unique keys.

## 4. Candidate Key

* A candidate key is a set of columns that can uniquely identify a record.
* A table can have multiple candidate keys, but only one is chosen as the primary key.

## 5. Composite Key

* A composite key is a combination of two or more columns that uniquely identify a record. It is used when a single column cannot guarantee uniqueness.

## 6. Super Key

* A super key is a set of one or more columns that can uniquely identify a record.
* It includes candidate keys and additional attributes that may not be necessary for uniqueness.
