# **Pandas :**

* Pandas is an open-source Python library used for data manipulation, analysis and cleaning.

### What is Pandas Used for?
* Reading and writing data from various file formats like CSV, Excel and SQL databases.
* Cleaning and preparing data (handling missing values, filtering, removing duplicates).
* Merging, joining, and reshaping datasets.
* Performing statistical analysis and descriptive statistics.

## **Data Structures in Pandas :**
* Pandas provides two data structures for manipulating data which are as follows:

### **1. Series :**

* A Series is a one-dimensional labeled array capable of holding any data type (integers, strings, floating-point numbers, Python objects, etc.).

* It’s similar to a column in a spreadsheet or a database table.

### **2. DataFrame :**

* Pandas DataFrame is a two-dimensional data structure with labeled axes (rows and columns).

* It is created by loading the datasets from existing storage which can be a SQL database, a CSV file or an Excel file.

* It can be created from lists, dictionaries, a list of dictionaries etc.
