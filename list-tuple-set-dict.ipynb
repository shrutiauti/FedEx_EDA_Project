# 1. List

* A list in python is a collection of items that are ordered and changeable . list are written with square brackets.
* fruits = ['apple','banana','cherry']
* You can add , remove or change items in a list using functions like append(), remove() or by using indexes like fruits[1]='orange'
* List can contain different data types - numbers,string even other list.
* They are useful when you want to store multiple items in one variable.


## Characteristics of a list
* Ordered
* Changeble/Mutable
* Hetrogeneous
* Can have duplicates
* are dynamic
* can be nested
* items can be accessed
* can contains any kind of objects in python

## Creating List

```python
# Empty 
print([])

# 1D List - homo
print([1,2,3,4,5])

# 2D List - hetro
print([1,2,3,4,[1,2,3]])

# 3D List
print([[[1,2],[3,4]],[[5,6],[7,8]]])

# Hetrogeneous
print([1,True,5.6,'Hello'])

# Using type conversion

print(list('hello'))
```

## Accessing Items from a list 

### 1. Indexing
* Indexing refers to the process of acessing specific element in a sequence by its position.
* Positive Indexing : uses non negative integer to access elements from the beginning of the sequence.
* Negative Indexing : uses negative integer to access elements from end of the sequence.
* basic syntax is : object[number]

### 2. Slicing
* Slicing in a python lets you extract a portion of a list , string or tuple using range of indexes.
* basic syntax is : object[start : stop : step]

#### Indexing Examples

```python
L = [1,2,3,6,7]
L2 = [1,2,3,[6,7]]
L3 = [[[1,2],[3,4]],[[5,6],[7,8]]]

# positive indexing

print(L[0])  # 1
print(L[2])  # 3
# print(L[6])  #IndexError: list index out of range
print(L2[3][0]) # 6

print(L3[1][0][1]) # 6
print(L3[0][1][0]) #3
print(L3[0][0][1]) #2
```

```python
# negative indexing

print(L[-1])  # 7
print(L2[-1][-2])  # 6
```

#### Slicing Examples

```python
L = [1,2,3,6,7]

print(L[0:2])

print(L[-3:-1])

print(L[-3:])

print(L[::2])

print(L[::-1])
```

### Adding Items to a list

### 1. Append()
* The Append() method adds a single element which can be String, Integer or another list to end of the list.
* If we pass a list or another iterable it will add the entire list as a single element.
* Syntax : list.append(element)

```python
L = [1,2,3,4,5]
L.append(6)
L.append([7,8,9])
L
```

### 2. Extend()
* The Extend method adds all the element of an iterable (list,tuple,string) to end of the list .
* unlike append() which adds a single element extend add each element from iterable to the list.
* syntax : list.extend(iterable)

```python
L = [1,2,3,4,5]
L.extend([6,7,8])
L.extend('java')
print(L)
```

### 3. Insert()
* The Insert method allows you to insert an element (String, object, integer) at any specified position in the list.
* if the index is greater than the list length, the element is added to the end. if index is negative it start counting from end of the list.
* Syntax : list.insert(index,element)

```python
L = [1,2,3,4,5]
L.insert(0,100)
L.insert(4,3.5)
L
```

### List Comprehension
* List Comprehension provides concise way to create new list based on values of an existing list.
* new_list = [ expression for item in iterable if condition == True]
* advantage : more time efficient & space efficient than loops , requires fewer lines of code.

```python
# Add 1 to 10 numbers to a list 

L = []

for i in range(1,11):
    L.append(i)

print(L)
```

```python
L = [ i for i in range(1,11)]
L
```

```python
# scalar multiplication on a vector

v = [2,3,4]
s = -3

[ s*i for i in v]
```

```python
# Add squares 

L = [1,2,3,4,5]

[i*i for i in L]
```

```python
# print all the numbers which are divisible by 5 in the range of 1 to 50

[ i for i in range(1,51) if i % 5 == 0]
```

```python
# find languages which start with letter p
languages= ['java','python','php','c','javascript']
[ i for i in languages if i.startswith('p')]
```

```python
# nested if with list comprehension

basket = ['apple','guava','cherry','banana']
my_fruits=['apple','kiwi','grapes','banana']

[i for i in my_fruits if i in basket if i.startswith('a')]
[i for i in my_fruits if i in basket and i.startswith('a')]

# add new list from my_list and items if the fruits exists in basket and also start with 'a'
```

```python
# prin a (3,3) matrix using  list comprehension -> nested list comprehension
[[i*j for i in range(1,4)] for j in range(1,4)]
```

```python
# cartesian products -> List comprehension on 2 lists together

L1 = [1,2,3,4]
L2 = [5,6,7,8]

[i*j for i in L1 for j in L2]
```

### Packing arguments *
*  packing is a technique used to group multiple values into single variable using * (for tuple or list) and ** (for dictionary)
*  *args (non-keyword arguments): packs multiple positional arguments into a tuple
*  **kwargs (keyword argument) : packs multiple keyword arguments into dictionary

```python
def sample(*args):
    print("packed args :",args)

sample(1,2,3,4,'pythonn')
```

* The function takes any number of arguments.
* The *args packs all arguments into tuple

### unpacking arguments 
* unpacking is used to extract individual values from a packed variable and assign them to seperate variables.

```python
# unpacking with *
def addition(a,b,c):
    return a + b + c

num = (1,4,6)
result= addition(*num) # *num unpacks the tuple into seperate arguments a,b,c
print(result)
```

```python
# unpacking with **
def info(name,age,country):
    print(f"Name :{name}, Age: {age}, country:{country}")

data={"name":"Ram", "age":23, "country":"India"}
info(**data)
```

```python
# packing and unpacking together 

def together(*args,**kwargs):
    print("positional :", args)
    print("keyword arguments :", kwargs)

together(1,2,3,name='ram',age=23)
```
